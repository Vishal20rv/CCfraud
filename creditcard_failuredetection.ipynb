```python
df = pd.read_csv('dataset.csv')
df
```

```python
df.nunique()#to get an unique no of data in columnn of dataset
```

```python
df.describe() #its for some statistical knowledge about numerical data columns
```

```python
df=df.drop(["trans_date_trans_time","dob","merch_lat","merch_long","trans_num"],axis='columns') #we drop these items becoz these seems unnecessary
```

```python
category_colu=[]#finding categorical data
for i in df.columns:
  if (df[i].dtype=='object'):
    category_colu.append(i)
print(category_colu)
```

```python
df=pd.get_dummies(df,columns=category_colu) #dumies for convert category datas into numerical format
df.describe()
```

```python
X=df.drop('is_fraud',axis=1)
Y=df['is_fraud']
```

```python
stdscaler=StandardScaler()#convert data into range 0-1
scaled_df=stdscaler.fit_transform(X)
scaled_df
```

```python
pca=PCA(n_components=20)#converting many colums into small columns of data
xpca=pca.fit_transform(scaled_df)
xpca.shape
```

```python
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test = train_test_split(xpca, Y,
test_size=0.33, random_state=42)
logreg = LogisticRegression(random_state = 0)
logreg.fit(X_train, y_train)
```

```python
Y_pred = logreg.predict(X_test)
print(Y_pred)
```

```python
accuracy_score(y_test, Y_pred)
```
